## Removing inconsistencies if any

```python
# Converting Date dataype from Object to Datetime
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
data
```

```python
# Adding a column for month of the date.
data['Month']=data['Date'].dt.month
data.head()
```

```python
data['Month'].unique()
```

```python
# Dropping the columns 'New' and 'PendingS' since they are having completely null values. And fulfilled-by also, since it has most NaN values.
# Dropping 'currency' and 'ship-country' columns since they are INR and IN respectively,which is not required for the analysis.
data=data.drop(columns=['New','PendingS','fulfilled-by','currency','ship-country'])
data
```

```python
# checking if any duplicated rows present in the dataset.
data.duplicated().sum()
```

There are 168 duplicates rows present.

```python
#removing those duplicates
data.drop_duplicates(inplace=True)
data=data.reset_index(drop=True)
data
```

```python
def state_edit(state):
    if state in ['Rajshthan', 'Rajasthan', 'Rajsthan', 'rajasthan', 'rajsthan', 'RJ']:
        return 'RAJASTHAN'
    elif state in ['Odisha', 'Orissa','orissa']:
        return 'ODISHA'
    elif state == 'Goa' or state == 'goa':
        return 'GOA'
    elif state in ['Delhi','New Delhi','delhi']:
        return 'DELHI'
    elif state == 'Chandigarh':
        return 'CHANDIGARH'
    elif state in ['Punjab', 'PB','punjab','Punjab/Mohali/Zirakpur']:
        return 'PUNJAB'
    elif state == 'Manipur':
        return 'MANIPUR'
    elif state in ['Arunachal Pradesh', 'AR','Arunachal pradesh','APO']:
        return 'ARUNACHAL PRADESH'
    elif state in ['Bihar', 'bihar']:
        return 'BIHAR'
    elif state == 'Mizoram':
        return 'MIZORAM'
    elif state == 'Gujarat':
        return 'GUJARAT'
    elif state == 'Meghalaya':
        return 'MEGHALAYA'
    elif state == 'Sikkim':
        return 'SIKKIM'
    elif state in ['Nagaland', 'NL']:
        return 'NAGALAND'
    elif state in ['Puducherry', 'Pondicherry']:
        return 'PUDUCHERRY'
    return state  # Return the original state if no match is found

# Apply the function to the city_state column
data['ship-state'] = data['ship-state'].apply(state_edit)
data['ship-state'].unique()
```

```python
data["ship-city"]=data["ship-city"].str.upper()
data["ship-city"].unique()
```

```python
data[['ship-city',	'ship-state',	'ship-postal-code']].isnull().sum()
```

Ship-city, ship-state, ship-postal-code contain 33 null values. Since having no idea about the place of package delivery, it is better to remove these rows

```python
data = data.dropna(subset=['ship-city', 'ship-state', 'ship-postal-code'])
data=data.reset_index(drop=True)
data
```

```python
data['Amount']=data['Amount'].fillna(data['Amount'].median())
data['ship-postal-code']=data['ship-postal-code'].astype(int)
data
```

```python
data['Month']=data['Month'].replace({3:'March',4:'April',5:'May',6:'June'})
data['Month'].unique()
```

## **4. Customer Segmentation**

```python
# Segment customers based on their buying behaviour, location, and other relevant factors.
customer=data.groupby('Order ID').agg({'Amount':'sum','Qty':'sum','ship-city':'first','ship-state':'first','ship-postal-code':'first','B2B':'first'}).reset_index()
customer.head()
```

```python
# Segmentation using clustering
features=customer[['Amount','Qty']]
```

```python
#standardize the features
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_features=scaler.fit_transform(features)
```

```python
from sklearn.cluster import KMeans
inertia = []
for n in range(1, 11):
    kmeans = KMeans(n_clusters=n, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)
```

```python
#plot the elbow curve
sns.set(style='white')
plt.plot(range(1,11),inertia,marker='o',color='brown')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

The elbow in the plot is around 3 clusters. This suggests that using 3 clusters is a reasonable choice for segmenting the customers.

```python
optimal_clusters=3
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
customer['Cluster'] = kmeans.fit_predict(scaled_features)
customer.head()
```

```python
cluster_analysis=customer.groupby('Cluster').agg({'Amount':['mean','sum'],'Qty':['mean','sum'],'ship-city':'count','ship-state':'first','ship-postal-code':'first','B2B':'first'}).reset_index()
cluster_analysis.columns=['Avg Amount Spent', 'Total Amount Spent','Avg Quantity Purchased','Total Quantity Purchased', 'Customer Count','Representative City','Representative State','Representative Postal Code', 'B28']
cluster_analysis
```

```python
plt.figure(figsize=(10,4))
sns.set(style='white')
sns.scatterplot(x='Amount', y='Qty', hue='Cluster',data=customer, palette ='viridis')
plt.title('Customer Segments')
plt.xlabel('Total Amount Spent')
plt.ylabel('Total Quantity Purchased')
plt.legend(loc ='lower right')
plt.show()
```

**Findings and Recommendations**   

Cluster 0 - Low spenders with infrequent buyers.   
Offer discounts for first purchases and bundle deals to encourage more frequent buys.   
Introduce rewards for repeat purchases to boost frequency.

Cluster 1 - Moderate spenders and low frequent buyers. Better than Cluster 0 groups.  
Offer discounts on future purchases and introduce subscription models.  
Run seasonal promotions and personalized email offers to increase purchase frequency.    

Cluster 2 - High amount spenders and frequent buyers and buy more quantity.   
Provide exclusive rewards, early access to products, and personalized offers.
Cross-sell premium products and bulk deals.  
Offer dedicated customer service for a better experience.
